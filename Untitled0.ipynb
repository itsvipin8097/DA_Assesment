Importing dependency and loading dataset using pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file into a DataFrame
df = pd.read_excel(r'/content/DA Assignment.xlsx')

# Access the data in the DataFrame
# For example, you can print the first few rows
print(df.head())
```

a.	How are the efficiency numbers looking like? Can you share your views?

In below code, we read the Excel file into a DataFrame. Then, we calculate various efficiency metrics such as resolution time, average first reply time within business hours, and average requester wait time within business hours.

```python

# Calculate resolution time statistics
resolution_time_stats = df['Resolution time'].describe()

# Calculate average resolution time
average_resolution_time = df['Resolution time'].mean()

# Calculate average first reply time within business hours
average_first_reply_time = df['First reply time in minutes within business hours'].mean()

# Calculate average requester wait time within business hours
average_requester_wait_time = df['Requester wait time in minutes within business hours'].mean()

# Print the statistics and average values
print("Resolution Time Statistics:")
print(resolution_time_stats)
print("\nAverage Resolution Time: {:.2f} minutes".format(average_resolution_time))
print("Average First Reply Time within Business Hours: {:.2f} minutes".format(average_first_reply_time))
print("Average Requester Wait Time within Business Hours: {:.2f} minutes".format(average_requester_wait_time))
```

b.	Which groups are quick, slow etc. Create a table + relevant charts.

In below code, we group the data by "Group" and calculate the average resolution time for each group. The resulting DataFrame group_avg_resolution_time is then sorted in ascending order based on the average resolution time. We create a table using the sorted data and display it. Additionally, we create a bar chart to visualize the average resolution time per group.

```python

# Group the data by "Group" and calculate the average resolution time for each group
group_avg_resolution_time = df.groupby("Group")["Resolution time"].mean().reset_index()

# Sort the groups by average resolution time in ascending order
group_avg_resolution_time_sorted = group_avg_resolution_time.sort_values("Resolution time")

# Create a table of average resolution time per group
table = group_avg_resolution_time_sorted.rename(columns={"Resolution time": "Average Resolution Time"})

# Print the table
print(table)

# Create a bar chart of average resolution time per group
plt.figure(figsize=(8,5))
plt.bar(table["Group"], table["Average Resolution Time"])
plt.xlabel("Group")
plt.ylabel("Average Resolution Time")
plt.title("Average Resolution Time by Group")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

c.	What type of tickets are taking the most time to resolve? Create a table + relevant charts.

we group the data by "Manual Tagging of Categories [list]" and calculate the average resolution time for each category. The resulting DataFrame category_avg_resolution_time is then sorted in descending order based on the average resolution time. We create a table using the sorted data and display it. Additionally, we create a bar chart to visualize the average resolution time per category.

```python

# Group the data by "Manual Tagging of Categories [list]" and calculate the average resolution time for each category
category_avg_resolution_time = df.groupby("Manual Tagging of Categories [list]")["Resolution time"].mean().reset_index()

# Sort the categories by average resolution time in descending order
category_avg_resolution_time_sorted = category_avg_resolution_time.sort_values("Resolution time", ascending=False)

# Create a table of average resolution time per category
table = category_avg_resolution_time_sorted.rename(columns={"Resolution time": "Average Resolution Time"})

# Print the table
print(table)

# Create a bar chart of average resolution time per category
plt.figure(figsize=(10, 6))
plt.bar(table["Manual Tagging of Categories [list]"], table["Average Resolution Time"])
plt.xlabel("Category")
plt.ylabel("Average Resolution Time")
plt.title("Average Resolution Time by Category")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

In below code create different type of insight from the given data which includes.

Average Resolution Time by Group

Request Status Distribution

Satisfaction Score Distribution

Distribution of Resolution Time

```python


# Insight 1: Average Resolution Time by Group
avg_resolution_time_group = df.groupby('Group')['Resolution time'].mean().sort_values(ascending=False)

# Bar chart for average resolution time by group
plt.figure(figsize=(7, 4))
plt.bar(avg_resolution_time_group.index, avg_resolution_time_group)
plt.xlabel('Group')
plt.ylabel('Average Resolution Time')
plt.title('Average Resolution Time by Group')
plt.xticks(rotation=45)
plt.show()
```

```python
# Insight 2: Request Status Distribution

status_counts = df['Status'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Request Status Distribution')
plt.axis('equal')
plt.show()
```

```python
# Insight 3: Satisfaction Score Distribution

# Convert "Satisfaction Score" to string/object data type
df['Satisfaction Score'] = df['Satisfaction Score'].astype(str)

# Plot the histogram for Satisfaction Score
plt.figure(figsize=(8, 6))
plt.hist(df['Satisfaction Score'].dropna(), bins=10, edgecolor='red')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.title('Distribution of Satisfaction Score')
plt.show()
```

```python
# Insight 2: Distribution of Resolution Time
plt.figure(figsize=(8, 6))
plt.hist(df['Resolution time'], bins=20, edgecolor='green')
plt.xlabel('Resolution Time')
plt.ylabel('Frequency')
plt.title('Distribution of Resolution Time')
plt.show()
```
